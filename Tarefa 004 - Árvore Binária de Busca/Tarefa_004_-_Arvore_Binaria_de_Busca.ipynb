```python
import math

class Point2D:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

class Reg_Node:
    def __init__(self, pos=Point2D(), data=0):
        self.pos = pos
        self.data = data

class no_PR_Quadtree:
    def __init__(self, top_left=Point2D(0, 0), bot_right=Point2D(0, 0)):
        self.top_left = top_left
        self.bot_right = bot_right
        self.n = None
        self.is_leaf = True
        self.top_left_tree = None
        self.top_right_tree = None
        self.bot_left_tree = None
        self.bot_right_tree = None

    def insert(self, node):
        if node is None:
            return

        if not self.in_boundary(node.pos):
            return

        if abs(self.top_left.x - self.bot_right.x) <= 1 and abs(self.top_left.y - self.bot_right.y) <= 1:
            if self.n is None:
                self.n = node
            return

        if self.is_leaf and self.n is None:
            self.n = node
            return

        if self.is_leaf and self.n is not None:
            self.is_leaf = False
            self.insert(self.n)

        if self.top_left.x + self.bot_right.x / 2 >= node.pos.x:
            if self.top_left.y + self.bot_right.y / 2 >= node.pos.y:
                if self.top_left_tree is None:
                    self.top_left_tree = no_PR_Quadtree(self.top_left, Point2D((self.top_left.x + self.bot_right.x) / 2,
                                                                              (self.top_left.y + self.bot_right.y) / 2))
                print(f"Desci pela topLeftTree com o ponto: {node.data}")
                self.top_left_tree.insert(node)
            else:
                if self.bot_left_tree is None:
                    self.bot_left_tree = no_PR_Quadtree(Point2D(self.top_left.x, (self.top_left.y + self.bot_right.y) / 2),
                                                        Point2D((self.top_left.x + self.bot_right.x) / 2, self.bot_right.y))
                print(f"Desci pela botLeftTree com o ponto: {node.data}")
                self.bot_left_tree.insert(node)
        else:
            if self.top_left.y + self.bot_right.y / 2 >= node.pos.y:
                if self.top_right_tree is None:
                    self.top_right_tree = no_PR_Quadtree(Point2D((self.top_left.x + self.bot_right.x) / 2, self.top_left.y),
                                                         Point2D(self.bot_right.x, (self.top_left.y + self.bot_right.y) / 2))
                print(f"Desci pela topRightTree com o ponto: {node.data}")
                self.top_right_tree.insert(node)
            else:
                if self.bot_right_tree is None:
                    self.bot_right_tree = no_PR_Quadtree(Point2D((self.top_left.x + self.bot_right.x) / 2,
                                                                 (self.top_left.y + self.bot_right.y) / 2),
                                                          Point2D(self.bot_right.x, self.bot_right.y))
                print(f"Desci pela botRightTree com o ponto: {node.data}")
                self.bot_right_tree.insert(node)

    def search(self, p):
        if not self.in_boundary(p):
            return None

        if self.is_leaf and self.n is not None:
            return self.n

        if self.top_left.x + self.bot_right.x / 2 >= p.x:
            if self.top_left.y + self.bot_right.y / 2 >= p.y:
                if self.top_left_tree is None:
                    return None
                return self.top_left_tree.search(p)
            else:
                if self.bot_left_tree is None:
                    return None
                return self.bot_left_tree.search(p)
        else:
            if self.top_left.y + self.bot_right.y / 2 >= p.y:
                if self.top_right_tree is None:
                    return None
                return self.top_right_tree.search(p)
            else:
                if self.bot_right_tree is None:
                    return None
                return self.bot_right_tree.search(p)

    def in_boundary(self, p):
        return self.top_left.x <= p.x <= self.bot_right.x and self.top_left.y <= p.y <= self.bot_right.y


# Driver program
if __name__ == "__main__":
    root = no_PR_Quadtree(Point2D(0, 0), Point2D(8, 8))
    a = Reg_Node(Point2D(1, 1), 11)
    b = Reg_Node(Point2D(2, 5), 22)
    c = Reg_Node(Point2D(7, 6), 33)
    d = Reg_Node(Point2D(5, 5), 44)
    e = Reg_Node(Point2D(7, 5), 55)
    root.insert(a)
    root.insert(b)
    root.insert(c)
    print("Aqui vai dar split")
    root.insert(d)
    root.insert(e)
    print(f"Node a: {root.search(Point2D(1, 1)).data}")
    print(f"Node b: {root.search(Point2D(2, 5)).data}")
    print(f"Node c: {root.search(Point2D(7, 6)).data}")
    print(f"Node d: {root.search(Point2D(5, 5)).data}")
    print(f"Node e: {root.search(Point2D(7, 5)).data}")
    print(f"Non-existing node: {root.search(Point2D(6, 6))}")
```
